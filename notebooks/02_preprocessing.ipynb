```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split

df=pd.read_csv('data/Loan_default.csv')
print(df.shape)
```

```python
#separate features and target
x=df.drop(['Default','LoanID'], axis=1)
y=df['Default']

print(x.shape)
print(x.dtypes)
```

```python
#encode categorical columns. text into numbers
le = LabelEncoder()

cat_cols=['Education', 'EmploymentType', 'MaritalStatus', 'HasMortgage','HasDependents','LoanPurpose','HasCoSigner']

for col in cat_cols:
    x[col]=le.fit_transform(x[col])

print(x.dtypes)
```

What LabelEncoder did:
For example EmploymentType which had values like Full-time, Part-time, Self-employed, Unemployed — it converted them to 0, 1, 2, 3.

```python
#split into train and test sets
x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=42, stratify=y
)

print('Training set:', x_train.shape)
print('Test set:', x_test.shape)
print('\nTraining target distribution:')
print(y_train.value_counts(normalize=True))
```

test_size=0.2 — 80% of data goes to training, 20% to testing.

stratify=class imbalance 88% non-default, 12% default. random split might be unrepresentative. starify guarantee both train and test sets maintain the same 88/12 ratio as the original data
